```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
## Visualizing a 3d function
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# meshgrid builds 2d grid from 1d arrays
# contour takes a grid of x vals, a grid of y vals, and a grid of z fals
```

```python
plt.contour(X, Y, Z, colors='black');  # negative vals are dashed lines by default
```

```python
# can configure the color of lines and the number to draw

plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
# see other color maps with tab completion

plt.cm.
```

```python
# can switch to a filled contour plot for better viewing

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
# to reduce graininess, could increase the number of contours, but this gets inefficient for python
# instead, use plt.imshow() which interprets a 2d grid of data as an image

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
          cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```

```python
# can combine images too!
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
          cmap='RdGy', alpha=0.5)
plt.colorbar();
```
